```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('kayak.csv')
df.head()
```

```python
df.columns
```

# Top-5 destinations

```python
# top Warmest cities in the nex days
top_cities = df.groupby('city')['temp_max_kelvin'].mean().sort_values(ascending=False).head(5)
top_cities.index
```

```python
top_5_df = df[df['city'].isin(top_cities.index)]
top_5_df
```

```python
fig = px.scatter_geo(top_5_df,
                     lat='latitude',
                     lon='longitude',
                     scope='europe',  # Set scope to Europe
                     center={"lat": 46.2276, "lon": 2.2137},  # Center on coordinates of France
                     color='temp_max_kelvin',  # Color the points by max temperature
                     hover_name='city',  # Hover text will show city name
                     size='temp_max_kelvin',  # Size of the points will be based on max temperature
                     projection='natural earth',  # Map projection style
                     title='Top 5 Destinations According to Max Temperature')  # Title of the map

# If the points are not visible or are too small, adjust the size parameter
fig.update_traces(marker=dict(size=12))  # Example to set a fixed size

# If the map is not zooming correctly, make sure you have valid latitude and longitude values
#fig.update_geos(fitbounds="locations", visible=True)  # Set visible to True to show the map features
fig.update_geos(
    lataxis_range=[42, 51],  # Latitude range, adjust as needed
    lonaxis_range=[-5, 10],  # Longitude range, adjust as needed
    showcountries=True,
    showcoastlines=True   
)

# Show the figure
fig.show()
```

# Top-20 hotels

### Lets get the best hotels but lets make sure we only have 1 per city

```python
# top_hotels_df = df.groupby('hotel_name').agg({
#     'hotel_rating': 'mean',
#     'latitude': 'first',
#     'longitude': 'first'
# }).sort_values(by='hotel_rating', ascending=False).head(20)

# top_hotels_df
```

```python
top_hotels_df = (
    df.sort_values(by='hotel_rating', ascending=False)  # Sort the hotels by rating
    .drop_duplicates(subset='city')  # Keep only the top-rated hotel per city
    .groupby('city')  # Group by city
    .agg({
        'hotel_rating': 'first',  # Since it's already sorted, 'first' will give us the highest rating
        'hotel_name': 'first',  # Get the name of the top-rated hotel
        'latitude': 'first',  # Assuming latitude is the same for all hotels in the city, otherwise you'd need to handle this differently
        'longitude': 'first'  # Same assumption as latitude
    })
    .sort_values(by='hotel_rating', ascending=False)  # Sort the cities by the highest hotel rating
    .head(20)  # Take the top 20
)
```

```python
top_hotels_df.reset_index(inplace=True)
```

```python
top_hotels_df.head()
```

```python
fig = px.scatter_geo(top_hotels_df,
                     lat='latitude',
                     lon='longitude',
                     scope='europe',  # Set the scope to Europe
                     center={"lat": 46.2276, "lon": 2.2137},  # Center on coordinates of France
                     color='hotel_rating',  # Color the points by hotel rating
                     size='hotel_rating',  # Size of the points will be based on hotel rating
                     hover_name=top_hotels_df.index,  # Hover text will show hotel name
                     projection='natural earth',  # Map projection style
                     title='Top 20 Hotels by Rating')  # Title of the map

# Adjust the map to show more details
fig.update_geos(
    showcountries=True,  # Show country borders
    showcoastlines=True,  # Show coastlines
    showland=True,  # Show land
    landcolor='rgb(217, 217, 217)',  # Land color
    countrycolor='rgb(204, 204, 204)'  # Country border color
)

# Update map layout
fig.update_layout(
    geo=dict(
        lataxis=dict(range=[41.0, 51.0]),  # Latitude range for zoom
        lonaxis=dict(range=[-5.0, 10.0]),  # Longitude range for zoom
    )
)

# Show the figure
fig.show()
```
